```python
import pandas as pd 
import numpy as np
import scipy.sparse
from scipy import sparse
from sklearn.metrics.pairwise import cosine_similarity
```

```python
users = pd.read_csv("TestUsers.csv")
ratings_data = pd.read_csv("Masked_User_Rest_Ratings.csv")
del users["Unnamed: 0"]
```

```python
allUsers = set()
for i in users["Users"]:
    allUsers.add(i)
```

```python
user_ratings= dict()
for index, row in ratings_data.iterrows():
    if row["User"] in allUsers:
        data = [row["Restaurant"], row["Ratings"]]
        if row["User"] in user_ratings:
            user_ratings[row["User"]].append(data)
        else:
            user_ratings[row["User"]] = [data]
```

```python
new_df = df[:int(4755991/100)].pivot_table(index = 'User', columns ='Restaurant', values = 'Ratings').fillna(0)
```

```python
print(new_df)
```

```python
def standardize(row):
    new_row = (row - row.mean())/(row.max()-row.min())
    return new_row

df_std = new_df.apply(standardize).T
print(df_std)
```

```python
sparse_df = sparse.csr_matrix(df_std.values)
corrMatrix = pd.DataFrame(cosine_similarity(sparse_df),index=new_df.columns,columns=new_df.columns)
corrMatrix
```

```python
corrMatrix = new_df.corr(method='pearson')
corrMatrix.head(6)
```

```python
def get_similar(restaurant,rating):
    similar_score = corrMatrix[restaurant]*(rating-2.5)
    similar_score = similar_score.sort_values(ascending=False)
    #print(type(similar_ratings))
    return similar_score
```

```python
action_lover = [(1,5),(2,1),(10,1)]
similar_scores = pd.DataFrame()
for movie,rating in action_lover:
    similar_scores = similar_scores.append(get_similar(movie,rating),ignore_index = True)

similar_scores.head(10)
```

```python
newresult = similar_scores.sum().sort_values(ascending=False).head(10)
```

```python
results = dict()
for key, value in user_ratings.items():
    similar_scores = pd.DataFrame()
    for data in value:
        try:
            similar_scores = similar_scores.append(get_similar(data[0],data[1]),ignore_index = True)
        except:
            pass
        #similar_scores.sum().sort_values(ascending=False)
    newresult = similar_scores.sum().sort_values(ascending=False).head(10)
    results[key] = newresult.index.tolist()
```

```python
for key, value in results.items():
    print(str(key) + " : " + str(value))
```

```python
with open("recommendations.txt", 'w') as f:
    for key, value in results.items():
        f.write('%s:%s\n' % (key, value))
```
